## Muestreo de señales - Fourier


<div class="alert alert-block alert-info">
<b>1)</b> Se busca realizar una librería en Python que nos permita obtener $N$ muestras de las siguientes señales de período $T_0=1/f_0$, muestreadas a una frecuencia $f_s$ :

</div>


+ **Senoidal**

> $s(k)=A_0.sin(2\pi.k.\frac{ f_0}{f_s} + \phi_0),\space\space \forall k \in \Bbb{R} \wedge k \in (0, N-1)$

+ __Ruido__ incorrelado de distribución Gaussiana y parámetros $\mu$ y $\sigma^2$

***

+ (Opcional) __Cuadrada__ con ciclo de actividad $d \in (0,1)$

> $s(k)=A_0,\space\space \forall k \in (0, N.d ]$ 

> $s(k)=-A_0,\space \forall k \in (N.d, N-1)$

+ (Opcional) __Triangular__ con punto de simetría $d \in (0,1)$ y coeficientes $a_{ij} \in \Bbb{R}$

> $s(k)=a_{11}.k/f_s+a_{12},\space\space \forall k \in (0, N.d ]$ 

> $s(k)=a_{21}.k/f_s+a_{22},\space \forall k \in (N.d, N-1)$


Es decir que se podría invocar la señal que genere la senoidal de la siguiente manera:


```python

def generador_senoidal (fs, f0, N, a0=1, p0=0):
    """ 
    
    brief:  Generador de señales senoidal, con argumentos
    
    fs:     frecuencia de muestreo de la señal [Hz]
    N:      cantidad de muestras de la señal a generar
    f0:     frecuencia de la senoidal [Hz]
    a0:     amplitud pico de la señal [V]
    p0:     fase de la señal sinusoidal [rad]
    
    como resultado la señal devuelve:
    
    signal: senoidal evaluada en cada instante 
    tt:     base de tiempo de la señal
    """    

    # comienzo de la función
    
    ts = 1/fs # tiempo de muestreo 
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (N-1)*ts, N).flatten()
    
    signal = a0 * np.sin( 2*np.pi*f0*tt + p0 ) 
    
    
    # fin de la función
    
    return tt, signal
```

```python
# generador de ruidoincorrelado de distribución Gaussiana y parámetros 𝜇 y 𝜎2
def generador_ruido (media, varianza, fs, N):
    """ 
    
    brief:  Generador de ruido, con argumentos
    
    fs:     frecuencia de muestreo de la señal [Hz]
    N:      cantidad de muestras de la señal a generar
    media:     media del ruido
    varianza:     varianza del ruido
    
    como resultado la señal devuelve:
    
    signal: senoidal evaluada en cada instante 
    tt:     base de tiempo de la señal
    """    

    # comienzo de la función
    
    ts = 1/fs # tiempo de muestreo 
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (N-1)*ts, N).flatten()
    
    signal = np.random.normal(media, np.sqrt(varianza), size = N)
    
    
    # fin de la función
    
    return tt, signal
```

```python
#generador de onda cuadrada con amplitud A y ciclo de trabajo d
def generador_cuadrada (fs, N, A, d):
    """ 
    
    brief:  Generador de señales senoidal, con argumentos
    
    fs:     frecuencia de muestreo de la señal [Hz]
    N:      cantidad de muestras de la señal a generar
    A:      Amplitud
    d:      ciclo de trabajo
    
    como resultado la señal devuelve:
    
    signal: senoidal evaluada en cada instante 
    tt:     base de tiempo de la señal
    """    

    # comienzo de la función
    
    ts = 1/fs # tiempo de muestreo 
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (N-1)*ts, N).flatten()
    
    signal = sig.square(2*np.pi*tt,d) 
    
    
    # fin de la función
    
    return tt, signal
```

```python
#generador de onda Triangular con punto de simetría d
def generador_triangular (fs, N, A, d):
    """ 
    
    brief:  Generador de señales senoidal, con argumentos
    
    fs:     frecuencia de muestreo de la señal [Hz]
    N:      cantidad de muestras de la señal a generar
    A:      Amplitud
    d:      ciclo de trabajo
    
    como resultado la señal devuelve:
    
    signal: senoidal evaluada en cada instante 
    tt:     base de tiempo de la señal
    """    

    # comienzo de la función
    
    ts = 1/fs # tiempo de muestreo 
    
    # grilla de sampleo temporal
    tt = np.linspace(0, (N-1)*ts, N).flatten()
    
    signal = np.abs(sig.sawtooth(np.pi*tt+np.pi/2,d))
    
    
    # fin de la función
    
    return tt, signal
```

<b>2.a)</b> (Opcional) Implemente un algoritmo que calcule la transformada discreta de Fourier (DFT). Utilice las funciones generadoras de señales de 1) para evaluar y ejemplificar su funcionamiento.
</div>

```python
# Cálculo de dft por definición

def dft(x):
    N = len(x)
    
    n = np.arange(N)
    k = n.reshape([N, 1])
    exp = np.exp(-((2j * np.pi * n * k) / N))
    x_ft = np.dot(x, exp)

    return x_ft


##################
# a.1) Senoidal #
#################

a0 = 1 # Volts
p0 = 0 # radianes
f0 = 10   # Hz

# Insertar aquí el código para generar y visualizar la señal
##############################################################

tt,x = generador_senoidal (fs, f0, N, a0, p0)

spectrum= (2/N)*np.abs(dft(x))

frec=np.linspace(0.0,fs/2,N/2).flatten()

#importa la mitad del espectro y la doble barra es para forzar resultado entero
half = spectrum[:N//2]

#%% Presentación gráfica de los resultados
#nueva figura    
plt.figure(1)

plt.stem(frec,half)

plt.title('DFT de senoidal 10Hz')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud [V]')
    
# presentar una leyenda para cada tipo de señal
#axes_hdl = plt.gca()
    
# añadir leyenda
#axes_hdl.legend(line_hdls, [str(f0)+ ' Hz'], loc='upper right'  )

#mostrar figura

plt.show()
```

<b>2.b)</b> (Opcional) Compare el tiempo de ejecución de la DFT implementada en con la **FFT** (implementada en **scipy.signal**) en función del tamaño de la señal N. Grafique el tiempo de ejecución de ambos casos en función de N.
</div>

```python
#vector de muestras potencia de 2

N  = [16, 32, 64, 128, 256, 512, 1024, 2048] 


a0 = 1     # Volts
p0 = 0     # radianes
f0 = 4     # Hz

cantidad_puntos = len(N)

tiempos_dft = []

tiempos_fft = []

#frecuencia de muestreo para que me coincida en 1 seg
fs = [16, 32, 64, 128, 256, 512, 1024, 2048] 

#Para cada punto
for i in range(cantidad_puntos):
    #calculamos la senoidal
    tt, x = generador_senoidal(fs[i], f0, N[i], a0, p0)

    #calculamos el tiempo de la fft
    
    inicio1 = time.perf_counter()
    spectrum1 = (2/N[i])*np.abs(sc.fft(x))
    half1 = spectrum1[:N[i]//2] 
    tiempos_fft.append(time.perf_counter() - inicio1)
    
    #calculamos el tiempo de la dft
    
    inicio2 = time.perf_counter()
    spectrum2 = (2/N[i])*np.abs(dft(x))
    halfDFT = spectrum2[:N[i]//2]
    tiempos_dft.append( time.perf_counter() - inicio2)

#Grafico
plt.figure(1)
plt_fft = plt.plot(N,tiempos_fft, '-*b')
plt_dft = plt.plot(N,tiempos_dft, '-*r')

axes_hdl = plt.gca()
axes_hdl.legend(('tiempo FFT', 'tiempo DFT') )

plt.title('Comparación de tiempos de ejecución FFT vs DFT' )
plt.xlabel('Muestras')
plt.ylabel('Tiempo[s]')

plt.show()
```

<b>3.a)</b> Verifique el efecto del leakage para una senoidal de $f_0 = f_S/4 + f_D$ siendo $f_D = (0.01,\, 0.25,\, 0.5)$, es decir una frecuencia de desintonía respecto al bin $f_S/4$.
</div>

```python
#######################################
# Tu simulación que genere resultados #
#######################################
fs = 1024
N = 1024
fd = [0,0.01,0.25,0.5]

a0 = 1 # Volts
p0 = 0 # radianes
f0 = fs//4  # Hz
indice = 0
espectros = []
for f in fd:

    tt,x = generador_senoidal (fs, f0+f, N, a0, p0)

    spectrum= (2/N)*np.abs(sc.fft(x))

    frec=np.linspace(0.0,fs/2,N/2).flatten()

    #importa la mitad del espectro y la doble barra es para forzar resultado entero
    half = spectrum[:N//2]
    espectros.append(half)
    indice +=1
    #%% Presentación gráfica de los resultados
    #nueva figura    
    plt.figure(1)

    plt.stem(frec,half)

    plt.title('FFT de senoidal '+ str(f0+f) +' Hz' )
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Amplitud [V]')

    #mostrar figura
    plt.show()
```

```python
sumatoria = [0,0,0,0]
for a in range(indice):
    for elementos in range(N//2):
        if elementos != fs//4:
            sumatoria[a]+= np.abs(espectros[a][elementos]*espectros[a][elementos])
            
tus_resultados = [ ['$ \lvert X(f_0) \lvert$', '$ \lvert X(f_0+1) \lvert $', '$\sum_{i=F} \lvert X(f_i) \lvert ^2 $'], 
                   ['',                        '',                           '$F:f \neq f_0$'], 
                  [str(espectros[0][fs//4]), str(espectros[0][fs//4+1]), str(sumatoria[0])], # <-- acá debería haber numeritos :)
                  [str(espectros[1][fs//4]), str(espectros[1][fs//4+1]), str(sumatoria[1])], # <-- acá debería haber numeritos :)
                  [str(espectros[2][fs//4]), str(espectros[2][fs//4+1]), str(sumatoria[2])], # <-- acá debería haber numeritos :)
                  [str(espectros[3][fs//4]), str(espectros[3][fs//4+1]), str(sumatoria[3])], # <-- acá debería haber numeritos :)
                 ]
df = DataFrame(tus_resultados, columns=['Frecuencia central', 'Primer adyacente', 'Resto de frecuencias'],
               index=['$f_0$ \ expr. matemática', 
                      '', 
                      '$f_S/4$', 
                      '$f_S/4+0.01$', 
                      '$f_S/4+0.25$', 
                      '$f_S/4+0.5$'])
HTML(df.to_html())
```

<div class="alert alert-block alert-info">
<b>3.b)</b> Verifique qué ocurre si a la señal de $f_0 = f_S/4 + 0.5$ se le agregan ceros para prolongar su duración. Es decir si la señal tiene N muestras, agregue $M_j$ ceros siendo $M_j = ( \frac{N}{10},\, N,\, 10N)$.
</div>

```python
from numpy import zeros
from numpy import hstack
# Inserte aquí su código

fs = 1024
N = 1024
M = [0,N//10,N,10*N]

a0 = 1 # Volts
p0 = 0 # radianes
f0 = fs//4  # Hz
indice = 0
espectros = []
for m in M:

    tt,x = generador_senoidal (fs, f0+0.5, N, a0, p0)
    
    z = zeros(m)
    N2 = N+ 2*z.size
    #ceros por izquierda
    x = hstack((z,x))
    #ceros por derecha
    x = hstack((x,z))
    
    spectrum= (2/N2)*np.abs(sc.fft(x))

    frec=np.linspace(0.0,fs/2,N2/2).flatten()

    #importa la mitad del espectro y la doble barra es para forzar resultado entero
    half = spectrum[:N2//2]
    espectros.append(half)
    indice +=1
    #%% Presentación gráfica de los resultados
    #nueva figura    
    plt.figure(1)

    plt.stem(frec,half)

    plt.title('FFT: senoidal 256.5Hz + vector de ' + str(z.size) +' ceros'  )
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Amplitud [V]')

    #mostrar figura
    plt.show()  
```

**3.e)** Analice cuál es la frecuencia  $\hat{f}_0 = \mathop{arg\ max}_f \{\lvert X(f) \rvert \}$ a la que ocurre el máximo del espectro y en cuánto difiere de $f_0$ en términos porcentuales.

```python
f_aprox = [0,0,0,0]
N3 = [N,N+N//5,3*N,21*N]
#para cada espectro
for a in range(indice):
    f_max = []
    sum = 0
    #calculo el maximo
    maximo = max(espectros[a])
    #busco la ubicación del máximo
    for muestras in range(len(espectros[a])):
        if espectros[a][muestras] == maximo:
            f_max.append(muestras)
    #sumo las frecuencias a las que de dan los máximos
    for elem in f_max:
        sum += elem 
    #promedio y normalizo
    f_aprox[a] = (sum/len(f_max)) * fs/N3[a]


tus_resultados = [str((abs(f_aprox[0]-(f0+0.5))/(f0+0.5))*100), # <-- acá debería haber numeritos :)
                  str((abs(f_aprox[1]-(f0+0.5))/(f0+0.5))*100), # <-- acá debería haber numeritos :)
                  str((abs(f_aprox[2]-(f0+0.5))/(f0+0.5))*100), # <-- acá debería haber numeritos :)
                  str((abs(f_aprox[3]-(f0+0.5))/(f0+0.5))*100), # <-- acá debería haber numeritos :)
                 ]
df = DataFrame(tus_resultados, columns=['$e_\%$'],
               index=['0',
                      '$N/10$', 
                      '$N$', 
                      '$10N$'])
HTML(df.to_html())
```
